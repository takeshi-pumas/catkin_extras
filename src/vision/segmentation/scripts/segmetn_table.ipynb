```python
zs_no_nans=np.zeros((corrected['z'].shape))
for i in range(corrected['z'].shape[0]):
    for j in range(corrected['z'].shape[1]):
        if np.isnan(corrected['z'][i,j]):
            zs_no_nans[i,j]=1.0
            print ('reject')
        else:
            zs_no_nans[i,j]=corrected['z'][i,j]
```
